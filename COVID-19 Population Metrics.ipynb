We use the time series data for daily new cases information.

```python
import pandas as pd

url="https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_confirmed_US.csv"
df=pd.read_csv(url)
```

```python
df
```

Let's get the total number of new cases per day in the US.

```python
a=df.groupby(['Country_Region']).sum()
a
```

```python
a_T=a.T
a_T
```

Here we get the information on total daily new cases in the US according to the dates.

```python
cases_confirmed=a_T.tail(174)
cases_confirmed.rename(columns={'Country_Region': 'Date', 'US': 'Cases_Confirmed'}, inplace=True)
```
